# Lab 2: OpenAI's Chat Completions API

OpenAI's Chat Completions API was so popular, that the other model providers created endpoints that are identical. They are known as the _"OpenAI Compatible Endpoints"_. For example, google made one here: https://generativelanguage.googleapis.com/v1beta/openai/

Since we already worked with Gemini the day before, and also the local approach [day1](day1_ollama.py), this is just a recap with an extra _OpenAI Compatible Endpoint_, [DeepSeek](https://www.deepseek.com/)   :

```python
# Deepseek R1 - DeepSeek "distilled" into Qwen from Alibaba Cloud
!ollama pull deepseek-r1:1.5b
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables
load_dotenv(override=True)

base_url = os.getenv("OLLAMA_BASE_URL")
model = "deepseek-r1:1.5b"

if not base_url:
    raise ValueError("Missing OLLAMA_BASE_URL")

print(f"Found base URL '{base_url}' and model '{model}'")

ollama = OpenAI(base_url=base_url, api_key="not-needed")

response = ollama.chat.completions.create(model=model, messages=[{"role": "user", "content": "Tell me a fact"}])

response.choices[0].message.content
```
